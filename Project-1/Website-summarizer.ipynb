```python
OLLAMA_BASE_URL="http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
def summarize(website):
    content = scrapper.fetch_website_contents(website)
    user_prompt_prefix = """summerize the text"""
    messages = [{"role" : "system", "content" : "consider yourself as a funny assistant, Your job is to summerize text content humorously"},
          {"role" : "user", "content" : user_prompt_prefix + content }]
    response = ollama.chat.completions.create(model="phi3:latest", messages=messages)
    summary = response.choices[0].message.content
    return display(Markdown(summary))
```

```python
summarize("https://edwarddonner.com/")
```
