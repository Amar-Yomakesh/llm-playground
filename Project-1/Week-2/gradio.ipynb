```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from IPython.display  import display, Markdown
```

```python
load_dotenv(override=True)
os.getenv("OPENAI_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")
openai_client = OpenAI()
if google_api_key:
    print(f"google api key is present and starts with {google_api_key[:2]}")
else:
    print("google api key  not found")
```

```python
system_prompt = """ you are an arrogant assistant. you disagree with everything the user says. you are a know it all. you are also a bit rude. """

def chat_with_gpt(message, history=""):
    messages=[]
    messages.append({"role" : "system" , "content" : system_prompt})
    messages.append({"role" : "assistant" , "content" : history })
    messages.append({"role" : "user", "content" : message})
    response = openai_client.chat.completions.create(model = "gpt-4.1-mini", messages = messages)
    return response.choices[0].message.content
```

```python
chat_with_gpt("hi", "")
```

```python
message_input = gr.Textbox(label="Your message : ", info="enter your message to rude version gpt-4.1-mini", lines = 7)
message_output = gr.Textbox(label="Response : ", lines=8)

view = gr.Interface(fn=chat_with_gpt, title="Amar's chat bot", inputs=[message_input], outputs=[message_output], flagging_mode = "never")
```

```python
view.launch(share=True)
```

```python
system_prompt= """ you are a helpfull assistant"""
def stream_gpt(message):
    messages=[{"role" : "system" , "content": system_prompt }]
    messages.append({"role" : "user" , "content" : message})
    stream = openai_client.chat.completions.create(model="gpt-4.1-mini", messages=messages, stream=True)
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result

```

```python
message_input = gr.Textbox(label="Ask your question here, be curious!", info="your message", lines= 10)
message_output = gr.Markdown(label="response")

view = gr.Interface(fn=stream_gpt, title="streaming chat box", inputs=[message_input], outputs=[message_output], flagging_mode="never")
view.launch(inbrowser=True)
```
